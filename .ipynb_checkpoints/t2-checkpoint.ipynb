```python
import sympy as sp
import numpy as np
def f(x:int, y:int):
    return (x-1)**2+100*(x**2-y)**2
f(3,2)
```

```python
x, y = sp.symbols('x y')
fun=(x-1)**2+100*(x**2-y)**2
def rad(fun):
    gr_dx=sp.diff(fun,x)
    gr_dy=sp.diff(fun,y)
    return np.array([gr_dx,gr_dy])
rad(fun)
```
