```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
 
def f(x):
    return x[0]**2 + x[0]*x[1] + x[1]**2
 
def gradient(x):
    return np.array([
    	2*x[0] + x[1],
    	x[0] + 2*x[1]
    ])
 
def gradient_method(x0, alpha, iterations):
    x = np.array(x0, dtype=float)
    resultats = []
    debut = time.time()
 
    for k in range(iterations):
    	grad = gradient(x)
    	d = -grad
    	x_new = x + alpha*d
 
    	resultats.append([
        	k,
        	x.copy(),
        	x_new.copy(),
        	np.linalg.norm(d),
        	np.linalg.norm(x_new-x),
        	np.linalg.norm(grad)
    	])
 
    	x = x_new
 
    fin = time.time()
    temps = fin - debut
 
    return resultats, temps
 
x0 = [2,2]
 
resultats1, temps1 = gradient_method(x0,1,100)
 
resultats2, temps2 = gradient_method(x0,0.5,300)
print(temps1)
print(temps2)
```
