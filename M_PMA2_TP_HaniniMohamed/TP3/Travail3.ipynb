```python
import numpy as np
import time

def gradient_method(x0, h, itmax, tol=1e-3):
    x = np.array(x0, dtype=float)
    resultats = []

    debut = time.time()

    for k in range(itmax):
        grad = gradient(x)
        d = -grad
        x_new = x + h * d

        resultats.append([
            k,
            x.copy(),
            x_new.copy(),
            np.linalg.norm(d),
            np.linalg.norm(x_new - x),
            np.linalg.norm(grad),
            f(x)
        ])

        if np.linalg.norm(x_new - x) < tol:
            break

        x = x_new

    fin = time.time()
    temps = fin - debut

    return resultats, temps
def f(x):
    return (x[0]**2 + x[0]*x[1] + x[0]*x[2]
        	+ x[1]*x[2] + x[1]**2 + x[2]**2)
 
def gradient(x):
    return np.array([
    	2*x[0] + x[1] + x[2],
    	x[0] + 2*x[1] + x[2],
    	x[0] + x[1] + 2*x[2]
    ])
 
x0 = [1,1,1]
 
resultats1, temps1 = gradient_method(x0,1,100)
 
resultats2, temps2 = gradient_method(x0,0.5,300)
print(temps1)
print(temps2)
```
