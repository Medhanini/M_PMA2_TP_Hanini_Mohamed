```python
import numpy as np
import matplotlib.pyplot as plt
import time

def f(x):
    return (x[0] - 1)**2 + 2 * x[1]**2

def gradient(x):
    return np.array([
        2 * (x[0] - 1),
        4 * x[1]
    ])

def gradient_method(x0, alpha, iterations):
    x = np.array(x0, dtype=float)
    resultats = []
    debut = time.time()

    for k in range(iterations):
        grad = gradient(x)
        d = -grad
        x_new = x + alpha * d

        resultats.append([
            k,
            x.copy(),
            x_new.copy(),
            np.linalg.norm(d),
            np.linalg.norm(x_new - x),
            np.linalg.norm(grad)
        ])

        x = x_new

    fin = time.time()
    return resultats, fin - debut

def gradient_method_tol(x0, alpha, tol, max_iter=10000):
    x = np.array(x0, dtype=float)
    k = 0
    debut = time.time()

    while np.linalg.norm(gradient(x)) > tol and k < max_iter:
        x = x - alpha * gradient(x)
        k += 1

    fin = time.time()
    return x, k, fin - debut

# Point initial
x0 = [0, 0]

# a) alpha = 1
resultats1, temps1 = gradient_method(x0, 1, 100)

# b) alpha = 0.5
resultats2, temps2 = gradient_method(x0, 0.5, 300)

print("Temps alpha = 1 :", temps1)
print("Temps alpha = 0.5 :", temps2)
```
