```python
import numpy as np
import matplotlib.pyplot as plt
import time

def f(v):
    x, y = v
    return 5*x**4 + 6*y**4 - 6*x**2 + 2*x*y + 5*y**2 + 15*x - 7*y + 13

def gradient(v):
    x, y = v
    return np.array([
        20*x**3 - 12*x + 2*y + 15,
        24*y**3 + 2*x + 10*y - 7
    ])

def gradient_method_tol(x0, alpha, tol, max_iter=10000):
    x = np.array(x0, dtype=float)
    k = 0
    debut = time.time()

    while np.linalg.norm(gradient(x)) > tol and k < max_iter:
        x = x - alpha * gradient(x)
        k += 1

        # Arrêt si débordement numérique
        if not np.all(np.isfinite(gradient(x))):
            break

    fin = time.time()
    return x, k, fin - debut

# Point initial
x0 = [0, 0]

# a) alpha = 1/2
x_a, k_a, t_a = gradient_method_tol(x0, 0.5, 1e-5)

# b) alpha = 1/4
x_b, k_b, t_b = gradient_method_tol(x0, 0.25, 1e-5)

# Affichage des résultats
print("===== a) α = 1/2 =====")
print("Solution :", x_a)
print("f(x) =", f(x_a))
print("||grad f(x)|| =", np.linalg.norm(gradient(x_a)))
print("Itérations :", k_a)
print("Temps :", t_a, "secondes")

print("\n===== b) α = 1/4 =====")
print("Solution :", x_b)
print("f(x) =", f(x_b))
print("||grad f(x)|| =", np.linalg.norm(gradient(x_b)))
print("Itérations :", k_b)
print("Temps :", t_b, "secondes")
```
