```python
import numpy as np
import matplotlib.pyplot as plt
def rosenbrock(x,y):
    return (x-1)**2 + 100*(x**2 - y)**2
def gradient_rosenbrock(x, y):
    dfdx = 2*(x - 1) + 400*x*(x**2 - y)
    dfdy = -200*(x**2 - y)
    return np.array([dfdx, dfdy])
def gradient_normalise(x, y, h=0.05, nit=200):
    xs = [x]; ys = [y]
    for k in range(nit):                   # exactement 200 iterations
        g = gradient_rosenbrock(x, y)
        longueur = np.sqrt(g[0]**2 + g[1]**2)
        x = x - h*g[0]/longueur            # pas de longueur h exactement
        y = y - h*g[1]/longueur
        xs.append(x); ys.append(y)

    return x, y, k

x, y, k = gradient_normalise(-1, 1)
print(x, y, k)
```
