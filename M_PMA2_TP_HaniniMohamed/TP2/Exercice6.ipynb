```python
import numpy as np
import matplotlib.pyplot as plt
def rosenbrock(x,y):
    return (x-1)**2 + 100*(x**2 - y)**2
def gradient_rosenbrock(x, y):
    dfdx = 2*(x - 1) + 400*x*(x**2 - y)
    dfdy = -200*(x**2 - y)
    return np.array([dfdx, dfdy])
def gradient_pas_1k(x, y, itmax=20):
    xs = [x]; ys = [y]
    for k in range(1, itmax + 1):          # k = 1, 2, 3, ...
        g = gradient_rosenbrock(x, y)
        if not np.all(np.isfinite(g)):     # inf ou nan : ca a explose, on stoppe
            break
        alpha = 1/k                        # le pas de ce tour
        x = x - alpha*g[0]
        y = y - alpha*g[1]
        xs.append(x); ys.append(y)
        print(f"k={k}  alpha={alpha:.3f}  x=({x:.4g}, {y:.4g})  f={rosenbrock(x, y):.4g}")
    return x, y, k

gradient_pas_1k(-1, 1)
```
