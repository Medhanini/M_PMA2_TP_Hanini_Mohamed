```python
import numpy as np
import matplotlib.pyplot as plt
def gradient_rosenbrock(x, y):
    dfdx = 2*(x - 1) + 400*x*(x**2 - y)
    dfdy = -200*(x**2 - y)
    return np.array([dfdx, dfdy])
    
def gradient_pas_fixe(x, y, h, tol=1e-3, itmax=100000):
    xs = [x]
    ys = [y]
    for k in range(itmax):
        g = gradient_rosenbrock(x, y)
        if np.sqrt(g[0]**2 + g[1]**2) <= tol:
            break
        x = x - h*g[0]
        y = y - h*g[1]
        xs.append(x)
        ys.append(y)
    plt.plot(xs, ys)
    plt.show()
    return x, y, k
x, y, k = gradient_pas_fixe(-1, 1, h=1e-3)
for h in [1e-4, 1e-3, 1.9e-3, 2.5e-3]:
    gradient_pas_fixe(-1, 1, h)
```
