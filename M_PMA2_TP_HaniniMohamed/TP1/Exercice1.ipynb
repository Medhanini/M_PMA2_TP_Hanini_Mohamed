```python
# --- Importation des packages ----------------------------------------
import numpy as np
from scipy.optimize import linprog
import math
import matplotlib.pyplot as plt

# --- 1) Création et indexation d'une matrice NumPy -------------------
A = np.array([[1, 2], [3, 4]])
print(A[0, 0])        # élément ligne 0, colonne 0 -> 1
print(A[:, 1])        # colonne d'indice 1         -> [2 4]
print(A[0, :])        # ligne d'indice 0           -> [1 2]
nl, nc = A.shape      # dimensions (lignes, colonnes)
print(nl, nc)         # -> 2 2

# --- 2) Type, dimensions, minimum et maximum -------------------------
c = np.array([[4, -2, -2]])
print(type(c))        # <class 'numpy.ndarray'>
print(c.ndim)         # 2 (tableau à 2 dimensions)
m = np.amin(c)
print(m)              # -2 : valeur minimale
print(np.argmin(c))   # 1 : indice (aplati) du premier minimum
n = np.amax(c)
print(np.argmax(c))   # 0 : indice (aplati) du premier maximum

# --- 3) arange, reshape et zeros --------------------------------------
a = np.array([[1.2, 2.5], [3.2, 1.8], [1.1, 4.3]])
v = np.arange(0, 10).reshape(2, 5)   # 0..9 réorganisés en 2 x 5
print(v)
print(a.reshape(3, 2))               # a est déjà 3 x 2 : inchangé
print(np.zeros(shape=(2, 4)))        # matrice 2 x 4 de zéros

# --- 4) Ajouter, insérer, supprimer des lignes/colonnes ---------------
b = np.array([[4.1, 2.6]])
c = np.append(a, b, axis=0)          # ajoute la ligne b en bas de a
print(c)
d = np.array([[7.8], [6.1], [5.4]])
print(np.append(a, d, axis=1))       # ajoute la colonne d à droite
print(np.insert(a, 1, b, axis=0))    # insère b en position 1
print(np.delete(a, 1, axis=0))       # supprime la ligne d'indice 1

# --- 5) Extraction de sous-tableaux (slicing) --------------------------
v = np.array([[1.2, 2.5], [3.2, 1.8], [1.1, 4.3]])
print(v[0:2, :])          # lignes 0 et 1
print(v[-1, :])           # dernière ligne
print(v[-2:, :])          # deux dernières lignes
print(np.max(v, axis=0))  # maximum de chaque colonne

# --- 6) Parcours d'un tableau et calcul d'une somme --------------------
s = 0
for i in range(0, v.shape[0]):
    for j in range(0, v.shape[1]):
        print(v[i, j])
        s = s + v[i, j]
print("Somme = ", s)      # 14.100000000000001 (arrondi flottant)

# --- 7) Définition d'une fonction : racine carrée ----------------------
def racine(x):
    return math.sqrt(x)

print(racine(4))          # 2.0

# --- 8) Les listes Python ----------------------------------------------
l = list()
l.append(2)
print(l)                  # [2]

# --- 9) Nuage de points : y = sin(x) ------------------------------------
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.figure()
plt.scatter(x, y, marker='o')

# --- 10) Nuage de points aléatoires (couleurs, tailles, colorbar) --------
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.figure()
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()

plt.show()
```
